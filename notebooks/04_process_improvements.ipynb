# Missed Refund Process Improvement Proposal

## Purpose

This notebook converts the findings from the exploratory analysis into practical recommendations for improving the Debt Held process.

The proposal focuses on reducing manual work, preventing outdated case information, improving prioritisation and creating a more reliable workflow for tracking missed refunds.

The recommendations are based on a synthetic dataset modelled on an existing operational process. No confidential customer or company data is included.

## Project Objective

The current process relies on a static Power BI snapshot that is manually exported into an Excel workbook.

The objective is to propose a future workflow that:

- keeps case information current
- reduces duplicate or unnecessary checks
- preserves staff tracking information
- prioritises higher-risk cases
- supports ongoing analysis and management reporting

## Current Debt Held Process

The current Debt Held process is largely manual and relies on a monthly Power BI snapshot.

Once the snapshot has been exported to Excel, agents manually review each case to determine whether a refund is still outstanding.

Because the workbook is based on a static export, it does not reflect changes made after the snapshot was taken. If a refund is processed before a case is reviewed, agents must still manually check the policy before confirming whether any action is required.

The process also relies on staff manually updating the workbook to record progress, making it difficult to maintain a live view of workload and outstanding cases.

## Problems Identified

The exploratory analysis identified several opportunities to improve the existing Debt Held process.

Some issues relate to operational efficiency, while others affect data quality, workload management and the ability to monitor performance over time.

The following recommendations are based on the findings from the exploratory analysis.

### 1. Reduce Static Snapshot Limitations

**Finding**

The current process relies on a static monthly Power BI snapshot. Once exported to Excel, the workbook does not reflect any refunds that are processed afterwards. Agents must therefore manually check every case to determine whether action is still required.

**Recommendation**

Introduce a scheduled refresh process so that the workbook is updated from a current data source rather than relying solely on the original monthly export.

Separating system-generated data from agent-entered tracking information would allow the latest refund status to be refreshed without overwriting manual notes or progress updates.

**Expected Benefits**

- Fewer unnecessary policy checks.
- Reduced duplicate work.
- Improved confidence that agents are reviewing current information.
- More accurate reporting of outstanding refunds.

### 2. Introduce Rule-Based Case Prioritisation

**Finding**

The current process allows agents to work through the workbook manually without a structured method of prioritising cases. This means cases may not always be reviewed in the most efficient order, and customers who have already contacted the business may wait alongside customers who are unaware of the outstanding refund.

**Recommendation**

Introduce an automated priority score based on operational business rules rather than refund value. For example:

- Customer has already contacted the business regarding the refund.
- Multiple policies exist for the same customer.
- Root cause indicates additional investigation may be required.
- Case has remained outstanding for an extended period.
- Cases approaching internal service level targets.

Cases could then be automatically sorted so that higher-priority work is presented consistently to agents.

**Expected Benefits**

- Improved customer experience.
- Reduced risk of repeat customer contact.
- More consistent workload management.
- A transparent and fair prioritisation process.

### 3. Improve Work Allocation

**Finding**

Currently, trained agents manually select cases from the shared workbook. As workload increases, this can lead to uneven distribution of work, duplicate effort and limited visibility of overall progress.

**Recommendation**

Introduce a structured case allocation process that assigns work consistently across trained agents. This could include:

- Assigning cases automatically or by team leader.
- Providing each agent with a filtered view of their allocated cases.
- Recording when a case is claimed to prevent duplicate working.
- Displaying live progress for managers and agents.

The process should remain flexible so work can be redistributed when priorities change or additional resource becomes available.

**Expected Benefits**

- Reduced duplicate working.
- More balanced workloads across agents.
- Improved visibility of progress.
- Faster completion of the monthly Debt Held process.

### 4. Monitor Performance Using Operational Reporting

**Finding**

The exploratory analysis showed that changes to operational processes, such as training, can influence the distribution of missed refund root causes over time.

However, the current monthly process provides limited visibility of trends until after the workload has been completed.

**Recommendation**

Develop a Power BI dashboard that monitors key performance indicators throughout the year, allowing managers to identify trends and measure the effectiveness of process improvements.

The dashboard could include:

- Total outstanding cases.
- Monthly case volumes.
- Root cause trends.
- Refund type distribution.
- Department breakdown.
- Outstanding refund value.
- Cases completed.
- Customer contact rates.

Monitoring these metrics over time would enable the business to identify emerging issues earlier and evaluate whether operational changes are delivering the intended improvements.

**Expected Benefits**

- Earlier identification of operational issues.
- Evidence-based decision making.
- Ability to measure the impact of training.
- Improved visibility for managers.

### 5. Use Root Cause Analysis to Support Continuous Improvement

**Finding**

The exploratory analysis identified recurring operational root causes, with payment date misunderstanding representing the largest contributor to missed refunds. The analysis also demonstrated how training influenced root cause trends over time.

**Recommendation**

Continue monitoring root causes through regular operational reporting to identify emerging trends and measure the effectiveness of process improvements.

Where recurring issues are identified, targeted interventions such as refresher training, process guidance or quality assurance reviews should be introduced and monitored over time.

**Expected Benefits**

- Reduced repeat errors.
- Improved effectiveness of training initiatives.
- Evidence-based quality assurance.
- Continuous operational improvement.

## Current workflow

```mermaid
flowchart TD

    A[Power BI Monthly Snapshot] --> B[Manual Export to Excel Workbook]

    B --> C[Agent Selects a Case]

    C --> D[Open Customer Policy]

    D --> E[Review Policy History and Payment Information]

    E --> F{Is a Refund Due?}

    F -->|Yes| G[Process Refund]

    F -->|No - Already Processed| H[Mark No Action]

    F -->|No - Refund Not Due| H

    G --> I[Add Policy Notes]

    I --> J[Update Excel Workbook]

    H --> J
```

## Proposed workflow

```mermaid
flowchart TD

    A[Customer Data Source] --> B[Scheduled Power Query Refresh]

    B --> C[Live Shared Case Tracker]

    C --> D[Automatic Case Prioritisation]

    D --> E[Case Allocated to Agent]

    E --> F[Open Customer Policy]

    F --> G[Review Policy History and Payment Information]

    G --> H{Is a Refund Due?}

    H -->|Yes| I[Process Refund]

    H -->|No - Already Processed| J[Mark No Action]

    H -->|No - Refund Not Due| J

    I --> K[Add Policy Notes]

    K --> L[Update Live Case Tracker]

    J --> L

    L --> M[Automatic Dashboard Refresh]

```

## Expected Business Benefits

Implementing the proposed improvements would provide benefits across operational efficiency, customer experience and management reporting.

The recommendations focus on improving the existing workflow rather than changing the underlying business process.

Expected benefits include:

- Reduced manual administration through automated data refreshes.
- Fewer unnecessary policy reviews caused by outdated snapshot data.
- More consistent workload allocation across trained agents.
- Improved visibility of outstanding work and process performance.
- Faster identification of operational issues through live reporting.
- Better customer outcomes by prioritising cases based on operational need.
- A scalable process capable of supporting increased workloads.

## Project Limitations

This proposal has been developed using a synthetic dataset based on an existing operational process.

While the business rules, workflow and analytical findings were designed to closely reflect the real Debt Held process, the project does not use live customer data or integrate directly with company systems.

The proposed future workflow should therefore be considered a proof of concept. Any real implementation would require collaboration with business stakeholders, access to operational systems and appropriate data governance.
